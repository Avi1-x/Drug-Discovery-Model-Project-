### **Read data**

```python
import pandas as pd

df_pre = pd.read_csv('Beta_Secretase_06_bioactivity_data_2class_pubchem_fp.csv', index_col=False)
can_smiles = pd.read_csv('canonicalsmiles.csv', index_col=False)
```

```python
df = pd.concat([df_pre,can_smiles], axis=1)
df
```

```python
y = df['class']
y
```

```python
X0 = df.drop(['class','canonical_smiles'], axis=1)
X0 = X0.fillna(0)
X0 = X0.astype('int64')
X = X0
X
```

### **Class distribution**

```python
y.value_counts()
```

```python
# Show pie plot (Approach 1)
y.value_counts().plot.pie(autopct='%.2f')
```

### **Random Undersampling**

``'not minority'`` = resample all classes but the minority class

```python
from imblearn.under_sampling import RandomUnderSampler

rus = RandomUnderSampler(sampling_strategy=1) # Numerical value
# rus = RandomUnderSampler(sampling_strategy="not minority") # String
X_res, y_res = rus.fit_resample(X, y)

ax = y_res.value_counts().plot.pie(autopct='%.2f')
_ = ax.set_title("Under-sampling")
```

```python
dropped_ids = [i for i in range(X.shape[0]) if i not in rus.sample_indices_]
ex_X = X.iloc[dropped_ids]
ex_X
```

```python
dropped_ids = [i for i in range(y.shape[0]) if i not in rus.sample_indices_]
ex_y = y.iloc[dropped_ids]  # for dataframes
ex_y
```

```python
df3 = pd.concat([ex_X, ex_y], axis=1)
df3.to_csv('Beta_Secretase_07_bioactivity_data_2class_pubchem_fp_external_set.csv', index=False)
df3
```
