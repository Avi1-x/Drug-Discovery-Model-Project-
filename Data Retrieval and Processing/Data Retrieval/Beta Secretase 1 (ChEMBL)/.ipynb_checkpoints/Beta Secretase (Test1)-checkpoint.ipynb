```python
# Import necessary libraries
import pandas as pd
from chembl_webresource_client.new_client import new_client
```

```python
# Target search 
target = new_client.target
target_query = target.search('Beta-Secretase')
targets = pd.DataFrame.from_dict(target_query)
targets
```

```python
selected_target = targets.target_chembl_id[2]
selected_target
```

```python
activity = new_client.activity
res = activity.filter(target_chembl_id=selected_target).filter(standard_type="IC50")
```

```python
df = pd.DataFrame.from_dict(res)
df
```

```python
df.to_csv('Beta_Secretase_01_bioactivity_data_raw.csv', index=False)
```

```python
df2 = df[df.standard_value.notna()]
df2 = df2[df.canonical_smiles.notna()]
df2
```

```python
len(df2.canonical_smiles.unique())
```

```python
df2_nr = df2.drop_duplicates(['canonical_smiles'])
df2_nr
```

```python
selection = ['molecule_chembl_id','canonical_smiles','standard_value']
df3 = df2_nr[selection]
df3
```

```python
df3.to_csv('Beta_Secretase_02_bioactivity_data_preprocessed.csv', index=False)
```

```python
df4 = pd.read_csv('Beta_Secretase_02_bioactivity_data_preprocessed.csv')
```

```python
bioactivity_threshold = []
for i in df4.standard_value:
  if float(i) >= 10000:
    bioactivity_threshold.append("inactive")
  elif float(i) <= 1000:
    bioactivity_threshold.append("active")
  else:
    bioactivity_threshold.append("intermediate")
```

```python
bioactivity_class = pd.Series(bioactivity_threshold, name='class')
df5 = pd.concat([df4, bioactivity_class], axis=1)
df5
```

```python
df5.to_csv('Beta_Secretase_03_bioactivity_data_curated.csv', index=False)
```
